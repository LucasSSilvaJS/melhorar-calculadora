```python
def soma(n1,n2):
    return n1+n2

def subtrai(n1,n2):
    return n1-n2

def multiplica(n1,n2):
    return n1*n2

def divide(n1,n2):
    return n1/n2

def potencia(n1,n2):
    return n1**n2

def menu():
    print('+====================+')
    print('| Menu de Alterações |')
    print('| 1 - somar          |')
    print('| 2 - subtrair       |')
    print('| 3 - multiplicar    |')
    print('| 4 - dividir        |')
    print('| 5 - potenciação    |')
    print('| 0 - sair           |')
    print('+====================+')

while True:
    menu()

    op = int(input('Selecione uma opção: '))
    
    if op not in (0,1,2,3,4,5):
        print('Opção inválida')
        continue

    if op == 0:
        print('Saindo...')
        break
    
    n1 = int(input('Digite o primeiro número: '))
    n2 = int(input('Digite o segundo número: '))
    
    match op:
        case 1:
            print(f'O resultado da soma de {n1} com {n2} é:', soma(n1,n2))
        case 2:
            print(f'O resultado da subtração de {n1} com {n2} é:', subtrai(n1,n2))
        case 3:
            print(f'O resultado da multiplicação de {n1} com {n2} é:', multiplica(n1,n2))
        case 4:
            print(f'O resultado da divisão de {n1} com {n2} é:', divide(n1,n2))
        case 5:
            print(f'O resultado da potencia de {n1} elevado a {n2} é:', potencia(n1,n2))
    
```
